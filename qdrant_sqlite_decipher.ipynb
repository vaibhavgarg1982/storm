```python
import sqlite3
import pickle

# Connect to the SQLite database
conn = sqlite3.connect(
    r"D:\storm\vector_store\vector_store\collection\my_documents\storage.sqlite"
)  # Adjust path as needed
cursor = conn.cursor()

# Query the points table
cursor.execute("SELECT id, point FROM points limit 100")
rows = cursor.fetchall()

for row in rows:
    point_id = row[0]
    point_data = row[1]

    # Deserialize the point data using pickle
    try:
        point_object = pickle.loads(point_data)
        print(f"{point_object.payload['metadata']['title']}")
        # print()
    except Exception as e:
        print(f"ID: {point_id}, Error: {e}")
    # break

# Close the connection
conn.close()
```
